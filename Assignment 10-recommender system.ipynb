```python
import numpy as np
import pandas as pd 
```

```python
books_df = pd.read_csv('books.csv', encoding ='unicode_escape')
books_df = books_df.drop(['Unnamed: 0'], axis=1)
books_df
```

```python
books_df[0:5]
```

```python
#number of unique users in the dataset
len(books_df.UserID.unique())
```

```python
len(books_df.BookTitle.unique())
```

```python
userbooks_df= books_df.pivot_table(index='UserID', columns='BookTitle', values='BookRating', 
               aggfunc='sum')
```

```python
userbooks_df
```

```python
userbooks_df.index = books_df.UserID.unique()
```

```python
userbooks_df
```

```python
#Impute those NaNs with 0 values
userbooks_df.fillna(0, inplace=True)
```

```python
userbooks_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( userbooks_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books_df.UserID.unique()
user_sim_df.columns = books_df.UserID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
